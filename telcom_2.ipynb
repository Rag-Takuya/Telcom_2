```python
# Carga el archivo CSV con los datos tratados
import pandas as pd

try:
    df = pd.read_csv('datos_tratados.csv') # Reemplaza con el nombre de tu archivo
    print("Datos cargados exitosamente.")
except FileNotFoundError:
    print("Error: El archivo 'datos_tratados.csv' no fue encontrado. Asegúrate de que el archivo esté en la ubicación correcta.")
    exit() # Salir si el archivo no se encuentra

# Elimina columnas que no aportan valor (ej. ID de cliente)
# Reemplaza 'ID_Cliente' con el nombre real de la columna de ID si existe
if 'ID_Cliente' in df.columns:
    df = df.drop('ID_Cliente', axis=1)
    print("Columna 'ID_Cliente' eliminada.")
else:
    print("La columna 'ID_Cliente' no se encontró para eliminar.")

# Transforma variables categóricas a formato numérico (One-Hot Encoding)
# Identifica automáticamente las columnas de tipo 'object' (cadenas) para codificar
categorical_cols = df.select_dtypes(include=['object']).columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True) # drop_first=True para evitar la trampa de las variables dummy
print(f"Variables categóricas codificadas: {list(categorical_cols)}")

# Calcula la proporción de clientes que cancelaron (asumiendo que la columna objetivo se llama 'Cancelacion')
# Reemplaza 'Cancelacion' con el nombre real de tu columna objetivo
if 'Cancelacion' in df.columns:
    cancel_proportion = df['Cancelacion'].value_counts(normalize=True) * 100
    print("\nProporción de clientes que cancelaron:")
    print(cancel_proportion)

    # Evalúa el desbalance de clases
    if cancel_proportion.min() < 20: # Umbral de ejemplo para considerar desbalance
        print("\nAdvertencia: Existe un posible desbalance significativo en las clases.")
        print("Considera aplicar técnicas de balanceo (undersampling, oversampling, SMOTE) si es necesario para tus modelos predictivos.")
else:
    print("\nAdvertencia: La columna 'Cancelacion' (o la columna objetivo) no se encontró para analizar el balance de clases.")


# Nota sobre normalización/estandarización:
# La necesidad de normalizar o estandarizar los datos depende de los modelos de machine learning que se aplicarán.
# Modelos como KNN, SVM, Regresión Logística y Redes Neuronales suelen requerirlo.
# Modelos basados en árboles (Decision Tree, Random Forest, XGBoost) generalmente no son sensibles a la escala.
# Considera este paso antes de entrenar tus modelos si usas algoritmos sensibles a la escala.

# Mostrar las primeras filas del dataframe procesado
print("\nPrimeras filas del dataframe después del preprocesamiento:")
display(df.head())
```

## Dividir los datos


Separar el dataframe en conjuntos de entrenamiento y prueba.


```python
# Importa la función necesaria
from sklearn.model_selection import train_test_split

# Define las características (X) y la variable objetivo (y)
X = df.drop('Churn', axis=1)
y = df['Churn']

# Divide los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print("Datos divididos en conjuntos de entrenamiento y prueba.")
print(f"Forma de X_train: {X_train.shape}")
print(f"Forma de X_test: {X_test.shape}")
print(f"Forma de y_train: {y_train.shape}")
print(f"Forma de y_test: {y_test.shape}")
```

## Escalar los datos


Aplicar estandarización o normalización a las características de los conjuntos de entrenamiento y prueba, ya que los modelos mencionados son sensibles a la escala de los datos.


```python
# Importa el escalador estándar
from sklearn.preprocessing import StandardScaler

# Instancia el StandardScaler
scaler = StandardScaler()

# Ajusta y transforma los datos de entrenamiento
X_train_scaled = scaler.fit_transform(X_train)

# Transforma los datos de prueba
X_test_scaled = scaler.transform(X_test)

print("Datos escalados exitosamente.")
```

# **ENTRENAR Y EVALUANR MODELOS**

### Entrenar y evaluar un modelo KNN.


```python
# Importa el clasificador KNN y métricas de evaluación
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Instancia y entrena el modelo KNN
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_scaled, y_train)

# Realiza predicciones
y_pred_knn = knn_model.predict(X_test_scaled)

# Calcula las métricas de evaluación
accuracy_knn = accuracy_score(y_test, y_pred_knn)
precision_knn = precision_score(y_test, y_pred_knn)
recall_knn = recall_score(y_test, y_pred_knn)
f1_knn = f1_score(y_test, y_pred_knn)

# Imprime las métricas
print("Métricas de Evaluación del Modelo KNN:")
print(f"Exactitud (Accuracy): {accuracy_knn:.4f}")
print(f"Precisión (Precision): {precision_knn:.4f}")
print(f"Sensibilidad (Recall): {recall_knn:.4f}")
print(f"Puntuación F1 (F1-score): {f1_knn:.4f}")
```

### Entrenar y evaluar un modelo SVM.


```python
# Importa el clasificador SVM y métricas de evaluación
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Instancia un modelo SVC
# Usando un kernel lineal por simplicidad, se puede experimentar con 'rbf' y C
svm_model = SVC(kernel='linear', random_state=42)

# Entrena el modelo SVM
svm_model.fit(X_train_scaled, y_train)

# Realiza predicciones en los datos de prueba escalados
y_pred_svm = svm_model.predict(X_test_scaled)

# Calcula las métricas de evaluación
accuracy_svm = accuracy_score(y_test, y_pred_svm)
precision_svm = precision_score(y_test, y_pred_svm)
recall_svm = recall_score(y_test, y_pred_svm)
f1_svm = f1_score(y_test, y_pred_svm)

# Imprime las métricas de evaluación
print("Métricas de Evaluación del Modelo SVM:")
print(f"Exactitud (Accuracy): {accuracy_svm:.4f}")
print(f"Precisión (Precision): {precision_svm:.4f}")
print(f"Sensibilidad (Recall): {recall_svm:.4f}")
print(f"Puntuación F1 (F1-score): {f1_svm:.4f}")
```

### Entrenar y evaluar un modelo de Regresión Logística.


```python
# Importa el modelo de Regresión Logística y métricas de evaluación
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Instancia un modelo de Regresión Logística
logreg_model = LogisticRegression(random_state=42)

# Entrena el modelo de regresión logística
logreg_model.fit(X_train_scaled, y_train)

# Realiza predicciones en los datos de prueba escalados
y_pred_logreg = logreg_model.predict(X_test_scaled)

# Calcula las métricas de evaluación
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
precision_logreg = precision_score(y_test, y_pred_logreg)
recall_logreg = recall_score(y_test, y_pred_logreg)
f1_logreg = f1_score(y_test, y_pred_logreg)

# Imprime las métricas de evaluación
print("Métricas de Evaluación del Modelo de Regresión Logística:")
print(f"Exactitud (Accuracy): {accuracy_logreg:.4f}")
print(f"Precisión (Precision): {precision_logreg:.4f}")
print(f"Sensibilidad (Recall): {recall_logreg:.4f}")
print(f"Puntuación F1 (F1-score): {f1_logreg:.4f}")
```

### Entrenar y evaluar un modelo de Red Neuronal simple.


```python
# Importa los módulos necesarios de tensorflow.keras
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.metrics import Precision, Recall

# Crea un modelo Sequential simple
nn_model = Sequential()

# Añade una capa Dense con activación ReLU
# La forma de entrada es el número de características en los datos de entrenamiento escalados
nn_model.add(Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)))

# Añade la capa de salida Dense con activación sigmoid para clasificación binaria
nn_model.add(Dense(1, activation='sigmoid'))

# Compila el modelo
nn_model.compile(optimizer='adam',
                 loss='binary_crossentropy',
                 metrics=['accuracy', Precision(), Recall()])

# Entrena el modelo de red neuronal
history = nn_model.fit(X_train_scaled, y_train,
                       epochs=10, # Número de épocas de entrenamiento
                       batch_size=32, # Tamaño del lote
                       validation_split=0.2) # Usa el 20% de los datos de entrenamiento para validación

# Evalúa el modelo entrenado en los datos de prueba escalados
print("\nEvaluación del Modelo de Red Neuronal:")
loss, accuracy_nn, precision_nn, recall_nn = nn_model.evaluate(X_test_scaled, y_test, verbose=0)

# Imprime los resultados de la evaluación
print(f"Pérdida (Loss): {loss:.4f}")
print(f"Exactitud (Accuracy): {accuracy_nn:.4f}")
print(f"Precisión (Precision): {precision_nn:.4f}")
print(f"Sensibilidad (Recall): {recall_nn:.4f}")

# Calcula la puntuación F1 para la red neuronal
# Maneja el caso donde precisión + sensibilidad es 0 para evitar la división por cero
f1_nn = 2 * (precision_nn * recall_nn) / (precision_nn + recall_nn) if (precision_nn + recall_nn) > 0 else 0
print(f"Puntuación F1 (F1-score): {f1_nn:.4f}")
```
